```python
# generalized probability function (see mathematica)
def probabilityGeneralized(xy, thetalaser, thetaBBO, thetabiref, phi, phil, A, C):
    x, y = xy
    anglex = x + thetaBBO
    angley = y + thetaBBO
    angle1 = thetabiref - thetalaser
    angle2 = thetaBBO - thetabiref
    term1 = np.cos(angle2)*(np.exp(1j*phi)*np.cos(angle1)*np.sin(anglex)*np.sin(angley) + np.cos(anglex)*np.cos(angley)*np.sin(angle1))
    term2 = np.exp(-1j*phil)*np.sin(angle2)*(np.exp(2j*phil)*np.cos(anglex)*np.cos(angley)*np.cos(angle1) - np.exp(1j*phi)*np.sin(anglex)*np.sin(angley)*np.sin(angle1))
    return A*np.abs(term1 + term2)**2 + C

# generalized probability but with thetalaser = 0 (see mathematica)
def probability(xy, thetaBBO, thetabiref, phi, phil, A, C):
    x, y = xy
    term1 = 1/2*(np.cos(x+thetaBBO)*np.cos(y+thetaBBO))*((1+np.exp(1j*phil))*np.sin(thetaBBO) + (np.exp(1j*phil)-1)*(np.sin(thetaBBO-2*thetabiref)))
    term2 = np.exp(1j*(phi-phil/2))*np.sin(x+thetaBBO)*np.sin(y+thetaBBO)*(np.cos(thetaBBO)*np.cos(phil/2) + 1j*np.cos(thetaBBO-2*thetabiref)*np.sin(phil/2))
    return A*np.abs(term1 + term2)**2 + C

# generalized probability but with thetalaser = 0, phi = 0, thetaBBO = thetabiref = 45 deg
def probabilitySimplified(xy, A, C):
    x, y = xy
    return A*1/2*(np.cos(x-y))**2 + C

# generalized probability but with phase and omega offsets
def probabilityGeneralized1(xy, thetalaser, thetaBBO, thetabiref, phi, phil, A, C, omegax, omegay, phasex, phasey):
    x, y = xy
    anglex = omegax*x + phasex + thetaBBO
    angley = omegay*y + phasey + thetaBBO
    angle1 = thetabiref - thetalaser
    angle2 = thetaBBO - thetabiref
    term1 = np.cos(angle2)*(np.exp(1j*phi)*np.cos(angle1)*np.sin(anglex)*np.sin(angley) + np.cos(anglex)*np.cos(angley)*np.sin(angle1))
    term2 = np.exp(-1j*phil)*np.sin(angle2)*(np.exp(2j*phil)*np.cos(anglex)*np.cos(angley)*np.cos(angle1) - np.exp(1j*phi)*np.sin(anglex)*np.sin(angley)*np.sin(angle1))
    return A*np.abs(term1 + term2)**2 + C

# probability but with phase and omega offsets
def probability1(xy, thetaBBO, thetabiref, phi, phil, A, C, omegax, omegay, phasex, phasey):
    x, y = xy
    x = omegax*x + phasex
    y = omegay*y + phasey
    term1 = 1/2*(np.cos(x+thetaBBO)*np.cos(y+thetaBBO))*((1+np.exp(1j*phil))*np.sin(thetaBBO) + (np.exp(1j*phil)-1)*(np.sin(thetaBBO-2*thetabiref)))
    term2 = np.exp(1j*(phi-phil/2))*np.sin(x+thetaBBO)*np.sin(y+thetaBBO)*(np.cos(thetaBBO)*np.cos(phil/2) + 1j*np.cos(thetaBBO-2*thetabiref)*np.sin(phil/2))
    return A*np.abs(term1 + term2)**2 + C

# simplified probability but with phase omega offsets
def probabilitySimplified1(xy, phase, omegax, omegay, A, C):
    x, y = xy
    return A*1/2*(np.cos(omegax*x-omegay*y-phase))**2 + C
```
